```python
import pandas as pd
```

```python
files = ['202005.csv', '202006.csv', '202007.csv']

list = []

for file in files:
    x = pd.read_csv(file)
    list.append(x)

df = pd.concat(list)

df
```

```python
df['gender'] = df['gender'].map({0: "Unknown", 1: "Male", 2: "Female"})
```

```python
df.to_csv('2020_data.csv', index=False)
```
